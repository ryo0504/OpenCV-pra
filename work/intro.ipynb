```python
def return_only_png(*args):
    png_images = []
    for arg in args:
        if arg.split(".")[1] == "png":
            png_images.append(arg)
    return png_images
```

```python
return_only_png("xxx.png", "yyy.jpg", "xxxx.png")
```
